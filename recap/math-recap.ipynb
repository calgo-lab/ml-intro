## Norm of a Vector

The $$p$$-**norm** $$\|\mathbf{v}\|_p$$ of a $$D$$-dimensional vector $$\mathbf{v}$$ is defined as

$$
\|\mathbf{v}\|_p = \left(\sum_i^D |v_i|^p \right)^{1/p}
$$

Usually (and in the following) $$\|\cdot\|$$ denotes the **Euclidean norm** ($$p=2$$):

$$
\|\mathbf{v}\|_2 = \left(\sum_i^D |v_i|^2 \right)^{1/2} = \sqrt{\sum_i^D v_i^2} = \sqrt{\mathbf{v}^{\top}\mathbf{v}}
$$

## Angles Between Vectors and Correlation Coefficients

Consider $$T$$ zero-mean$^*$ samples from two univariate time series $$x_i \in \mathbb{R}^1,\ y_i \in \mathbb{R}^1,\ i \in \{1, \dots, T\}$$ stored in vectors $$\mathbf{x},\ \mathbf{y}$$:

$$
\mathbf{x} = \begin{bmatrix} x_1 \\ \vdots \\ x_T \end{bmatrix} \qquad
\mathbf{y} = \begin{bmatrix} y_1 \\ \vdots \\ y_T \end{bmatrix}
$$

The **correlation coefficient** between $$\mathbf{x}$$ and $$\mathbf{y}$$ is defined as

$$
\frac{\sum_i^T x_i y_i}{\sqrt{\sum_i x_i^2}\,\sqrt{\sum_i y_i^2}} = \frac{\mathbf{x}^{\top}\mathbf{y}}{\|\mathbf{x}\|\,\|\mathbf{y}\|}
$$

$\Rightarrow$ **The cosine of the angle between $$\mathbf{x}$$ and $$\mathbf{y}$$ is their correlation!**

---
*$^*\sum_i^T x_i = \sum_i^T y_i = 0$*

## Axioms of Vector Operations

1. **Associativity of addition**
$$
\mathbf{u} + (\mathbf{v} + \mathbf{w}) = (\mathbf{u} + \mathbf{v}) + \mathbf{w}
$$

2. **Commutativity of addition**
$$
\mathbf{v} + \mathbf{w} = \mathbf{w} + \mathbf{v}
$$

3. **Distributivity of scaling**
$$
a(\mathbf{v} + \mathbf{w}) = a\mathbf{w} + a\mathbf{v}
$$

## Matrices

This is an example of a matrix $$A \in \mathbb{R}^{U \times V}$$:

$$
A = \begin{bmatrix}
A_{11} & A_{12} & \ldots & A_{1V} \\
A_{21} & A_{22} & \ldots & A_{2V} \\
\vdots & \vdots & \ddots & \vdots \\
A_{U1} & A_{U2} & \ldots & A_{UV}
\end{bmatrix}
$$

## Matrix Multiplication

Matrix $$A \in \mathbb{R}^{U \times D}$$ can be multiplied with a vector $$\mathbf{v} \in \mathbb{R}^{D}$$ by

$$
A\mathbf{v} =
\begin{bmatrix}
A_{11} & A_{12} & \ldots & A_{1D} \\
A_{21} & A_{22} & \ldots & A_{2D} \\
\vdots & \vdots & \ddots & \vdots \\
A_{U1} & A_{U2} & \ldots & A_{UD}
\end{bmatrix}
\begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_D \end{bmatrix}
=
\begin{bmatrix}
A_{11}v_1 + A_{12}v_2 + \ldots + A_{1D}v_D \\
A_{21}v_1 + A_{22}v_2 + \ldots + A_{2D}v_D \\
\vdots \\
A_{U1}v_1 + A_{U2}v_2 + \ldots + A_{UD}v_D
\end{bmatrix}
$$

## Matrices as Linear Mappings

A matrix $$A \in \mathbb{R}^{U \times V}$$ can be understood as a **linear map** from a $$V$$-dimensional space to a $$U$$-dimensional space.

| | |
|:---:|:---|
| ![Scalar product vTw](figures/math_prep_figures_vTw) | Consider the scalar product $$\mathbf{v}^{\top}\mathbf{w}$$ <br><br> $$\mathbf{w}$$ is mapped from $$\mathbb{R}^2 \rightarrow \mathbb{R}^1$$ <br><br> But the mapping $$\mathbf{v}^{\top} \in \mathbb{R}^{1 \times 2}$$ is a very simple case |

## Matrices as Linear Mappings — Types

- **Permutation of dimensions:**
$$A = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$$

- **Mirroring:**
$$A = \begin{bmatrix} -1 & 0 \\ 0 & -1 \end{bmatrix}$$

- **Scaling:**
$$A = \begin{bmatrix} a & 0 \\ 0 & b \end{bmatrix}$$

- **Rotation** by angle $$\alpha$$ in radians:
$$A = \begin{bmatrix} \cos(\alpha) & -\sin(\alpha) \\ \sin(\alpha) & \cos(\alpha) \end{bmatrix}$$

## Axioms of Matrix Operations

1. **Associativity**
$$
A(BC) = (AB)C
$$

2. **Distributivity**
$$
A(B + C) = AB + AC \qquad (A + B)C = AC + BC
$$

3. **No Commutativity!**
$$
AB \neq BA
$$

## Matrix Multiplication — Some Examples (Solutions)

$$
\begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}
\begin{bmatrix} 3 & 2 \\ 15 & -17 \end{bmatrix}
= \begin{bmatrix} 3 & 2 \\ 15 & -17 \end{bmatrix}
$$

$$
\begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}
\begin{bmatrix} 3 & 2 \\ -2 & 7 \end{bmatrix}
= \begin{bmatrix} -2 & 7 \\ 3 & 2 \end{bmatrix}
$$

$$
\begin{bmatrix} 2 & 3 \\ -1 & 5 \end{bmatrix}
\begin{bmatrix} 3 & 2 & 8 \\ 7 & 5 & 1 \end{bmatrix}
= \begin{bmatrix} 27 & 19 & 19 \\ 32 & 23 & -3 \end{bmatrix}
$$

## Taking Derivatives of Univariate Functions

| **Function** | **Derivative** |
|:---:|:---:|
| $$f(x) = x^n$$ | $$f'(x) = nx^{n-1}$$ |
| $$cf(x)$$ | $$cf'(x)$$ |
| $$f(x) + g(x)$$ | $$f'(x) + g'(x)$$ |
| $$f(x)g(x)$$ | $$f'(x)g(x) + f(x)g'(x)$$ |
| $$\dfrac{f(x)}{g(x)}$$ | $$\dfrac{f'(x)g(x) - f(x)g'(x)}{g(x)^2}$$ |
| $$f(g(x))$$ | $$f'(g(x))\, g'(x)$$ |

## Taking Derivatives of Vector-Valued Functions — Gradient

Consider a vector-valued function $$f(\mathbf{x}): \mathbb{R}^D \rightarrow \mathbb{R}^U,\ \mathbf{x} \in \mathbb{R}^D$$

> **The gradient $$\nabla f(\mathbf{x})$$ at position $$\mathbf{x}$$ is**
>
> $$
> \nabla f(\mathbf{x}) = \left[ \frac{\partial f(\mathbf{x})}{\partial x_1},\ \dots,\ \frac{\partial f(\mathbf{x})}{\partial x_D} \right]
> $$

## Taking Derivatives of Vector-Valued Functions — Rules

| **Function** | **Derivative** |
|:---:|:---:|
| $$f(\mathbf{x}) = \mathbf{u}^{\top}\mathbf{x}$$ | $$f'(\mathbf{x}) = \mathbf{u}$$ |
| $$f(\mathbf{x}) = \mathbf{x}^{\top}\mathbf{x}$$ | $$f'(\mathbf{x}) = 2\mathbf{x}$$ |
| $$f(\mathbf{x}) = A\mathbf{x}$$ | $$f'(\mathbf{x}) = A$$ |
| $$f(\mathbf{x}) = \mathbf{x}^{\top}A\mathbf{x}$$ | $$f'(\mathbf{x}) = \mathbf{x}^{\top}(A + A^{\top})$$ |
