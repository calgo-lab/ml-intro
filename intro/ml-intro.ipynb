## Machine Learning or Artificial Intelligence?

- Intelligence is very difficult to define ([Tomasello 2003](#references))
- Artificial and human intelligence is difficult to compare ([Turing 1950](#references))
- Scientists use a simpler term: Machine Learning (ML):
  - **=> ML are Algorithms that learn from data.**
- (since 2022, everyone calls it AI ... even researchers.)

## Machine Learning -- The Big Picture

Given data $x \in \mathcal{X}$ (neural signals, internet user data, ...) ML predicts a target variable $y \in \mathcal{Y}$ (thoughts, user intentions, ...)

> **ML algorithms learn a function $f(\cdot)$ from examples $(x, y)$**
> $
> f(x) = y
> $

Most of ML is about:
- Find the right function class for $f(\cdot)$
- Fitting $f(\cdot)$ correctly

### Machine Learning Pipelines

A full ML pipeline consists of **feature extraction** and the Machine Learning (ML) part.

![ML Pipeline](figures/ml-pipeline.png)

ML is mostly about fitting $f(x)$ but transforming your data (text, images, ...) into $d$-dimensional vectors $\vec{x} \in \mathbb{R}^d$ correctly is crucial.

### Machine Learning Paradigms

- **Supervised Learning:** If you have a lot of annotated data
  - Classification
  - Regression
- **Unsupervised Learning:** If you have data but not annotations
  - Classical unsupervised (Clustering, PCA, ...)
  - Semi-Supervised
  - Self-Supervised, Contrastive
  - Zero-Shot
- If you have a little data and like to annotate more
  - Active Learning
  - Few-Shot
- **Reinforcement Learning:** No data but task knowledge

### Machine Learning In Practice

### What is Cognition?

- Derived from *cognoscere* (latin): *to recognize*
- Refers to processing of information:
  - Learning, recognizing and categorizing concepts
  - Inference based on these concepts (e.g. language)
- Cognitive processes can be:
  - natural or artificial, conscious or unconscious
- **Easier to define and measure than intelligence**
- Cognition is studied by:
  - linguistics, neuroscience, philosophy, psychology
  - AND computer science (artificial intelligence, machine learning)

### Rule Based AI

- Rule based AIs can be very useful without learning from data
- Great example: ELIZA ([Weizenbaum 1966](#references))
- Weizenbaum used his domain expertise and knowledge about language structure to build a chat bot for psychotherapy
- **Advantages** of rule based systems:
  - Rules can be understood by humans
  - Testing the system is simple: system behaviour is deterministic
- **Disadvantages** of rule based systems:
  - Rules need to be written by humans
  - Rules need to be maintained (for example for new words)

### Rule Based AI: ELIZA ([Weizenbaum 1966](#references))

## Ideas of ML: Learning indepdendent of Sensory Modality

- Many algorithms from Cyberneticists are still state of the art
- Understanding biological cognition improves artificial cognition
- Cybernetic principles are independent of sensory modality!
- This is also a goal of Machine Learning:  
  Modeling of cognitive mechanisms (independent of data type).

> **Also in biological systems cognitive mechanisms can be decoupled from sensory modalities**
> - => Audiovisual re-routing in animals ([Roe et al. 1992](#references))
> - => Sensory substitution ([Bach-y-Rita 1969](#references))

### Sensory Substitution

Actually, no surgical rewiring is needed to show how the brain learns independent from sensory modalities (like ML algorithms and like cyberneticists postulated)

> **Sensory Substitution:**  
> Program new cognitive functions into existing sensory modality

Examples:

**Haptic Vision** ([Bach-y-Rita 1969](#references))  
=> Luminance on photosensors translated to vibrations on skin

**Feelspace** ([Nagel et al. 2005](#references))  
=> Allocentric orientation through haptic feedback

### Sensory Substitution: Tactile Vision

|---|---|
| ![Tactile vision device](figures/tactile-vision-1.pdf) <br> *Tactile vision device* <br><br> ![Experimental setup](figures/tactile-vision-2.pdf) <br> *Experimental setup* | Pioneered by [Bach-y-Rita 1969](#references), tested in [Jansson 1983](#references) <br><br> - Tactile matrix (20x20 vibrators) on the subject's back <br> - Tactile information about catching device and ball <br> - **=> Blind subjects can catch the ball!** |


### Sensory Substitution: Feelspace

And this not only works for lost modalities - it also works to acquire entirely new modalities!

| | |
|---|---|
| ![Feelspace belt](figures/feelspace_belt.pdf) <br> *Feelspace belt* <br> [http://feelspace.cogsci.uni-osnabrueck.de/](http://feelspace.cogsci.uni-osnabrueck.de/) | Creating an allocentric sense of orientation ([Nagel et al. 2005](#references)) <br><br> - Direction of north pole is translated to vibration in belt <br> - After wearing the belt for some weeks subjects acquire an allocentric sense of orientation <br><br> *"I was intuitively aware of the direction of my home or of my office."* |

## Summary

- **Cognitive processes:**  
  Perception, recognition of / inference on semantic concepts
- **Cybernetics**
  - studied cognitive processes inspired by biology
  - different from classical engineering approaches
  - modeled motion detection, navigation, ...
  - => but what about higher cognitive functions?
- **Artificial intelligence**
  - Old AI: rule based systems (like ELIZA)
  - AI Winter: Most research stopped, AI hype broke down
  - New AI (machine learning): learns rules from data
- **Risk and Potentials**
  - Higher degrees of automation
  - Questions around responsibility / accountability / fairness
  - Ethical dilemma around building and maintaining AI
